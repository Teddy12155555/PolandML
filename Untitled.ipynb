```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
filename = 'IRIS.csv' 
iris = pd.read_csv(filename, encoding='utf8')

#sns.relplot( x =  "sepal_length",y ="sepal_width", hue = 'species',data = iris)
newDF = pd.DataFrame(columns=['petal_length','sepal_width'])
newDF['petal_length'] = iris['petal_length'].copy()
newDF['sepal_width'] = iris['sepal_width'].copy()
Y = iris['species'].copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(newDF, Y, test_size=0.33)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#LDA Method

import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
clf = LinearDiscriminantAnalysis()
clf.fit(X_train, y_train)  
y_pred=clf.predict(X_test)
ytrain_pred=clf.predict(X_train)
names=iris['species'].unique()
N=len(names)
Output=np.zeros([N,N])
Output_Train=np.zeros([N,N])
for i in range(N):
    for j in range(N):
        Output[i,j]=sum((y_pred==names[i])&(y_test==names[j]))
        Output_Train[i,j]=sum((ytrain_pred==names[i])&(y_train==names[j]))
pd.DataFrame(Output,names,names)
pd.DataFrame(Output_Train,names,names)
```
