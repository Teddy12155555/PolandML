```python
import numpy as np
import matplotlib.pyplot as plt
# matplotlib inline
```

```python
x = np.linspace(0, 2*np.pi)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
y2 = np.cos(x)

#plt.plot(x,y, x,y2)
plt.plot(x, y, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.legend(loc='lower left')
plt.show()
```

### Histogram

```python
h = np.random.normal(1, 2, 100)

plt.subplot(1,2,1)
n, bins, patches = plt.hist(h, bins=50)

plt.subplot(1, 2, 2)
plt.hist(h, bins=20, histtype='bar', edgecolor='k', density=True)
plt.suptitle('Histogram')
plt.show()
```

```python
x = np.random.normal(0,2,100)
y = np.random.normal(0,2,100)

plt.figure()
plt.hist2d(x, y, bins=(20, 20), cmap='BuGn')
plt.colorbar()
plt.title('hist2d plot')

plt.figure()
plt.hexbin(x, y,bins=20, gridsize=20, extent=(-5, 5, -5, 5), cmap='Reds')
plt.title('hexbin plot')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(-4, 4,0.1)
y = np.arange(-4, 4,0.2)
X, Y = np.meshgrid(x, y)
Z=np.sin(X)+np.cos(Y)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.view_init(45, -45)
ax.plot_surface(X, Y, Z, cmap='terrain')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('2xy')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,np.pi*2,100)
y = np.sin(x)
plt.plot(x, y, '-.b', label='$sin(x)$')

x = np.linspace(0,np.pi*2,100)
y_ = np.cos(x)
plt.plot(x, y_, '-r', label='$cos(x)$',linewidth=7.0)

plt.fill_between(x, y_, y, color='green', alpha=0.2)

plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')

plt.title('Graph')
plt.legend(loc='lower left')
plt.grid(which='major', axis='y', linestyle='-')

plt.xticks([0, 0.5*np.pi,1*np.pi,1.5*np.pi,2*np.pi], ['$0$', '$0.5 \pi$','$1 \pi$','$1.5 \pi$','$2 \pi$'])
plt.yticks([-1,0,1])



plt.show()
```
