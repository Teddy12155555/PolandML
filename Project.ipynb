```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
dataset = pd.read_csv('googleplaystore.csv',encoding='utf8')
Rating=dataset['Rating']
Reviews=dataset['Reviews']
Size=dataset['Size']
Category=dataset['Category']
Installs=dataset['Installs']
df1=pd.DataFrame(Category)
df1['Rating']=Rating.interpolate(method ='linear', limit_direction ='forward') 
df1['Reviews']=Reviews
df1['Size']=Size
df1['Installs']=Installs
df1['Installs'].replace(to_replace='+',value='')
Zuicha=df1[df1['Rating']<1.5]
Jicha=df1[df1['Rating']<2.0][df1[df1['Rating']<2.0]['Rating']>=1.5]
Jiaocha=df1[df1['Rating']<2.5][df1[df1['Rating']<2.5]['Rating']>=2.0]
Cha=df1[df1['Rating']<3.0][df1[df1['Rating']<3.0]['Rating']>=2.5]
Hao=df1[df1['Rating']<3.5][df1[df1['Rating']<3.5]['Rating']>=3.0]
Jiaohao=df1[df1['Rating']<4.0][df1[df1['Rating']<4.0]['Rating']>=3.5]
Jihao=df1[df1['Rating']<4.5][df1[df1['Rating']<4.5]['Rating']>=4.0]
Zuihao=df1[df1['Rating']<5.0][df1[df1['Rating']<5.0]['Rating']>=4.5]
Lis=[Zuicha,Jicha,Jiaocha,Cha,Hao,Jiaohao,Jihao,Zuihao]
for i in Lis:
    print(i.groupby('Category').count())
#    print(i.count())
df1['Installs'] = df1['Installs'].str.replace('+', "")
df1['Installs'] = df1['Installs'].str.replace(',','')
df1=df1[df1['Installs']!='Free']
df1['Installs']= pd.DataFrame(df1['Installs'],dtype=float)
df1['Installs']=Rating.interpolate(method ='linear', limit_direction ='forward') 
df1['Installs']=np.log(df1['Installs'])
df1['Reviews']= pd.DataFrame(df1['Reviews'],dtype=float)
df1['Reviews']=Rating.interpolate(method ='linear', limit_direction ='forward') 
df1['Reviews']=np.log(df1['Reviews'])
df1['Size'] = df1['Size'].str.replace('M', "")
df1['Size']=Rating.interpolate(method ='linear', limit_direction ='forward') 
df1
#These above all is all about how to deal with our data
```
