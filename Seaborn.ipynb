## Copies, views and different methods of indexing arrays and dataframes

```python
array1=np.arange(5)
```

```python
array2=array1
```

```python
array1
```

```python
array2
```

```python
array2[3]=15 #Assignment on elements from array2 is reflected to array1 - array2 is not a distinct object
```

```python
array1
```

```python
array3=array2.copy() #.copy() method creates a new, distinct array
```

```python
array3
```

```python
array2
```

```python
array3[3]=0
```

```python
array3
```

```python
array2 #Assigning elements from copy do not change the original array
```

```python
array_slice=array2[1:4:2] #Slice produces a view of an array - not a copy
```

```python
array_slice
```

```python
array_slice[0]=100
```

```python
array_slice
```

```python
array2
```

```python
array_fancy=array2[[1,3]] #"Fancy" indexing produces a copy of an array
```

```python
array_fancy
```

```python
array_fancy[0]=200
```

```python
array_fancy
```

```python
array2
```

```python
sns.get_dataset_names() #Returns the list of names of all datasets stored in online Seaborn repository
```

```python
tips=sns.load_dataset('tips') #Loading dataset from seaborn online repository
```

```python
tips
```

```python
sns.relplot(x='total_bill',y='tip', data=tips) #Relation plot
```

```python
sns.relplot(x='total_bill',y='tip', hue='day', data=tips) #Relation plot with color of nodes differing according to the day of week
```

```python
sns.relplot(x='total_bill',y='tip', hue='day', style='time', data=tips) #Relation plot with style of nodes differing according to the time
```

```python
sns.catplot(x='day', y='tip', data=tips) #Category plot
```

```python
sns.catplot(x='sex', y='total_bill', data=tips, kind='swarm') #Category plot with points seperated in horizontal axis
```

```python
sns.boxplot(data=tips) #Boxplot for all columns (those which consists of numbers) of dataframe
```

```python
sns.boxplot(x='sex', y='total_bill', data=tips) #Boxplots for different categories (here determined by the 'sex' column)
```

```python
sns.catplot(x='day', y='tip', kind='bar', data=tips) #Barplot
```

```python
sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=True) #Relation plot with regression line added (by default fit_reg attribute is equal to True)
```

```python
x=np.random.normal(0,1,1000)
sns.distplot(x, kde=False, bins=30) #Histogram of an array - kde=False hides the density estimator plot, bins determines number of cells
```

```python
sns.kdeplot(x) #Kernel density estimator plot
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
sns.pairplot(data=iris, hue='species') #Relation plots for each pair of numerical variables (for two the same - kde plot), colors differ according to the species
```

```python
sns.pairplot(data=tips, hue='day')
```

```python
sns.jointplot(x='petal_length',y='sepal_length', data=iris) #Joint plot combining relation plot with histograms
```

```python
sns.relplot(y='tip',x='total_bill', hue='day', size='size', data=tips) #Sizes of nodes differs according to the 'size' variable
```

```python
#All data stored in IRIS dataset presented on one plot
palette=sns.diverging_palette(220,10,as_cmap=True)
sns.relplot(x='petal_width',y='petal_length', hue='sepal_length',palette=palette, size='sepal_width', style='species', data=iris)
```

```python
corr=iris.corr() #Creating dataframe consisting of correlation coefficients of columns of iris dataframe
```

```python
corr
```

```python
mask=np.zeros((4,4), dtype=np.bool) #Creating a mask to hide the upper traingle of the corr array
mask[np.triu_indices_from(mask)]=True
mask
```

```python
sns.heatmap(np.abs(corr),cmap=palette,mask=mask) #Presenting correlation coefficients on the heatmap, using mask to hide redunant or irrelevant cells
```
